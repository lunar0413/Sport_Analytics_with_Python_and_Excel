```python
df_ohtani["1B"] = df_ohtani["H"]-df_ohtani["2B"]-df_ohtani["3B"]-df_ohtani["HR"]
df_ohtani
```

```python
# Calculate the total bases
total_bases = (df_ohtani["1B"].sum() * 1 +
               df_ohtani["2B"].sum() * 2 +
               df_ohtani["3B"].sum() * 3 +
               df_ohtani["HR"].sum() * 4)

# Calculate the total number of at-bats
total_at_bats = df_ohtani["AB"].sum()

# Calculate the slugging percentage (SLG)
SLG_ohtani = total_bases / total_at_bats

# Print the slugging percentage
print(SLG_ohtani)
```

```python
# Calculate the sums
AB_ohtani = df_ohtani["AB"].sum()
H_ohtani = df_ohtani["H"].sum()
BB_ohtani = df_ohtani["BB"].sum()
HBP_ohtani = df_ohtani["HBP"].sum()
SF_ohtani = df_ohtani["SF"].sum()

# Calculate the on base percentage (OBP)
OBP_ohtani = (H_ohtani + BB_ohtani + HBP_ohtani)/(AB_ohtani + BB_ohtani + HBP_ohtani + SF_ohtani)

# Print the OBP
print(OBP_ohtani)
```

```python
# Adding OBP and SLG
OPS_ohtani = OBP_ohtani + SLG_ohtani

# Print the OPS
print(OPS_ohtani)
```

```python
# Calculating Runs Created using the variables created when calculating OBP and SLG
RC_ohtani = ((H_ohtani + BB_ohtani) * total_bases)/(AB_ohtani + BB_ohtani)

# Print the Runs Created
print(RC_ohtani)

# Per season
print(RC_ohtani/6)
```
